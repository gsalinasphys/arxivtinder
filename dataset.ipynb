```python
# Filtering dataset
import pandas as pd
from pathlib import Path
```

```python
# Divide dataset into chunks
chunk_size = 10_000
batch_no = 1

for chunk in pd.read_json('static/arxiv-metadata-oai-snapshot.json', lines=True, chunksize=chunk_size):
    chunk.to_json('static/chunks/chunk'+str(batch_no)+'.json')
    batch_no+=1
```

```python
# Reading a chunk of data from file
achunk = pd.read_json('static/chunks/chunk1.json')
```

```python
# Adding column with just the years
achunk['year'] = [int(date[:4]) for date in achunk['update_date']]
```

```python
achunk.head()
```

```python
from_yr = 2012
by_category = ['hep-ph', 'hep-th']

# Filter papers starting from year 'from_yr' and categories in list 'by_category' for a single chunk
achunk_filter = achunk.loc[(achunk['year']>=from_yr) & (achunk['categories'].isin(by_category))].copy()
achunk_filter.drop('year', axis=1, inplace=True)
```

```python
achunk_filter.head()
```

```python
# Filter papers starting from year 'from_yr' for the whole dataset
from_yr = 2017

directory = 'static/chunks/'
files = Path(directory).glob('chunk*.json')
frames = []
for file in files:
    achunk = pd.read_json(file)
    
    achunk['year'] = [int(date[:4]) for date in achunk['update_date']]
    
    achunk_filter = achunk.loc[achunk['year']>=from_yr].copy()
    achunk_filter.drop('year', axis=1, inplace=True)

    frames.append(achunk_filter)
```

```python
arxiv_from17 = pd.concat(frames)
```

```python
arxiv_from17.head()
```

```python
len(arxiv_from17)
```

```python
arxiv_from17.to_json('static/arxiv-metadata-oai-snapshot_from17.json')
```

```python
# Filter papers starting from year 'from_yr' and categories in list 'by_category' for the whole dataset
from_yr = 2012
by_category = ['hep-ph', 'hep-th']

directory = 'static/chunks/'
files = Path(directory).glob('chunk*.json')
frames = []
for file in files:
    achunk = pd.read_json(file)
    
    achunk['year'] = [int(date[:4]) for date in achunk['update_date']]
    
    achunk_filter = achunk.loc[(achunk['year']>=from_yr) & (achunk['categories'].isin(by_category))].copy()
    achunk_filter.drop('year', axis=1, inplace=True)

    frames.append(achunk_filter)
```

```python
arxiv_hepthph = pd.concat(frames)
```
